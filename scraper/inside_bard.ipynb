```python
from urllib.request import urlretrieve
import os
 
```

```python
import urllib.request
from urllib.parse import urlparse

host = "https://inside.bard.edu"
header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
}
rel_route = "/academic/courses/fall2024/menu.htm"
url = host + rel_route 



opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)
urllib.request.urlretrieve(url, f"./data/inside.html")
```

```python
html_file_path = "data/inside.html"

from bs4 import BeautifulSoup
with open(html_file_path, encoding="utf8") as fp:
    soup = BeautifulSoup(fp, 'html.parser')
```

```python
print(len(soup.find_all('a')))
subpage_prefix = "https://inside.bard.edu/academic/courses/current/"
url_pool = [ link.get('href') for link in soup.find_all('a') ] 
url_pool = [ url for url in url_pool if not url.startswith("http") ] 

for url in url_pool:
    try:
        urllib.request.urlretrieve(subpage_prefix + url, f"./data/{url}")
    except:
        pass 
    print(url)

```
